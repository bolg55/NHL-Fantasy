```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, TimeoutException
from bs4 import BeautifulSoup
import time
import csv
```

```python
options = Options()
options.headless = True # Run Firefox in headless mode
driver = webdriver.Firefox(options=options)
```

```python
try:
    # Navigate to the webpage
    driver.get('https://puckpedia.com/players/search')

    # Initialize WebDriverWait
    wait = WebDriverWait(driver, 20)

    # Wait for the "Get Started" button to become clickable
    get_started_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "Get Started")]')))
    get_started_button.click()

    # Wait for the page to fully load
    time.sleep(5)

    print("Page loaded. Looking for season dropdown...")

    # Get all divs with id='focus_season' and use the first one
    focus_season_divs = driver.find_elements(By.ID, "focus_season")
    print(f"Found {len(focus_season_divs)} season dropdowns")

    # Get the button inside the first dropdown div
    season_dropdown_button = focus_season_divs[0].find_element(By.TAG_NAME, "button")
    
    # Scroll to it and click using JavaScript
    driver.execute_script("arguments[0].scrollIntoView(true);", season_dropdown_button)
    time.sleep(1)
    print("Clicking dropdown button...")
    driver.execute_script("arguments[0].click();", season_dropdown_button)

    # Wait for dropdown to open
    time.sleep(2)

    # Debug: Check if options are visible
    print("Checking for dropdown options...")
    all_li_elements = driver.find_elements(By.TAG_NAME, "li")
    print(f"Found {len(all_li_elements)} li elements on page")
    
    # Look for options with data-value attribute
    season_options = driver.find_elements(By.XPATH, '//li[@data-value]')
    print(f"Found {len(season_options)} season options with data-value")
    
    if len(season_options) > 0:
        print("Sample options:")
        for opt in season_options[:5]:
            print(f"  - data-value: {opt.get_attribute('data-value')}, text: {opt.get_attribute('data-text')}")
    
    # Try to find the 2025-2026 option in multiple ways
    print("\nSearching for 2025-2026 option...")
    try:
        # Try method 1: by data-value
        season_option = driver.find_element(By.XPATH, '//li[@data-value="162"]')
        print("Found by data-value='162'")
    except:
        try:
            # Try method 2: by data-text
            season_option = driver.find_element(By.XPATH, '//li[@data-text="2025-2026"]')
            print("Found by data-text='2025-2026'")
        except:
            # Try method 3: by text content
            season_option = driver.find_element(By.XPATH, '//li[contains(text(), "2025-2026")]')
            print("Found by text content")
    
    print("Clicking 2025-2026 option...")
    driver.execute_script("arguments[0].click();", season_option)

    # Wait for the data to reload
    print("Waiting for data to reload...")
    time.sleep(4)
    wait.until(EC.presence_of_all_elements_located((By.XPATH, '//table/tbody/tr')))
    print("Data loaded with 2025-2026 season!")

    # Open a CSV file to write data
    with open('nhl_players_2025_26.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Player', 'Team', 'Position', 'Cap Hit', 'pv']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        has_next_page = True
        page_count = 0

        while has_next_page:
            page_count += 1
            print(f"Scraping page {page_count}...")
            
            # Wait for player data to load
            wait.until(EC.presence_of_all_elements_located((By.XPATH, '//table/tbody/tr')))

            # Find all the player rows
            rows = driver.find_elements(By.XPATH, '//table/tbody/tr')

            # Loop through each row to extract data
            for row in rows:
                try:
                    # Player's name
                    first_name = row.find_element(By.XPATH, './/span[@x-text="x.p_fn"]').text
                    last_name = row.find_element(By.XPATH, './/span[@x-text="x.p_ln"]').text
                    full_name = f'{first_name} {last_name}'

                    # Team name
                    team_element = row.find_element(By.XPATH, './/td[contains(@class, "flex justify-center")]//a')
                    team_href = team_element.get_attribute('href')
                    team_name = team_href.split('/')[-1].replace('-', ' ').title()

                    # Position
                    position = row.find_element(By.XPATH, './/td[@x-text="x.pos"]').text

                    # Salary Cap Hit
                    cap_hit = row.find_element(By.XPATH, './/td[@x-text="$store.puck.formatContractValue(x.cap_hit)"]').text

                    if cap_hit and cap_hit != 'N/A':
                        cap_hit_value = int(cap_hit.replace('$','').replace(',',''))
                        pv = round(cap_hit_value / 1_000_000, 3)
                    else:
                        cap_hit_value = 0
                        pv = 0
                            
                    # Write the data to CSV
                    writer.writerow({
                        'Player': full_name,
                        'Team': team_name if team_name != 'Utah Hc' else 'Utah Hockey Club',
                        'Position': position if position == 'D' else 'F',
                        'Cap Hit': cap_hit,
                        'pv': pv
                    })

                except Exception as e:
                    print(f"Error processing row: {e}")

            # Attempt to click the "Next" button
            try:
                # Scroll to the top of the page
                driver.execute_script("window.scrollTo(0, 0);")
                time.sleep(1)

                # Wait for the "Next" button
                next_button = wait.until(EC.presence_of_element_located((By.XPATH, '//button[.//i[contains(@class, "fa-angle-right")]]')))

                # Check if disabled
                next_button_class = next_button.get_attribute('class')

                if 'opacity-25' in next_button_class:
                    print("Reached last page!")
                    has_next_page = False
                else:
                    driver.execute_script("arguments[0].click();", next_button)
                    wait.until(EC.staleness_of(rows[0]))
                    time.sleep(3)
                    wait.until(EC.presence_of_all_elements_located((By.XPATH, '//table/tbody/tr')))

            except (NoSuchElementException, TimeoutException) as e:
                print(f"Error with Next button: {e}")
                has_next_page = False

    print(f"\nScraping completed! Scraped {page_count} pages. Check nhl_players_2025_26.csv")

finally:
    driver.quit()
```
